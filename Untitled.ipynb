```python
import requests
geoURL = 'https://opendata.arcgis.com/datasets/53ea466a189b4f43b3dfb7b38fa7f3b6_1.geojson'
r = requests.get(geoURL)

with open('boston.geojson', 'wb') as f:
    f.write(r.content)
```

```python
import folium
import numpy as np
import json

with open('boston.geojson') as e:
    boston_geo = json.load(e)

# Output: {'name': 'Bob', 'languages': ['English', 'Fench']}
print(boston_geo)
```
